### html에서 정보 추출

```python
from bs4 import BeautifulSoup

# 데이터 찾아주는 객체
# 데이터 탐색, 추출
indeed_soup = BeautifulSoup(indeed_result.text, "html.parser")
print(indeed_soup)
```

```python
# div를 앚아서 class가 pagination인걸 찾음
# div : division (section 분할한다고 생각)
pagination = indeed_soup.find("div", {"class":"pagination"})
print(pagination)
```

```python
# page : list
# a : anchor (hyper link)
# 링크를 모두 찾음
links = pagination.find_all('a')
print(links)
```

```python
for link in links:
    print(link.find("span"))
```

```python
pages = []
for link in links:
    pages.append(link.find("span"))
pages = pages[:-1]
print(pages) # 마지막꺼 제외하고 가져옴
```

```python
pages = []
for link in links:
    pages.append(link.find("span").string) # 그 안에 string만 가져오게 설정
pages = pages[:-1]
print(pages)
```

```python
# link(anchor)에서 string 추출해도 똑같은 결과가 나옴
# anchor가 있고, 이 요소 안에 다른 요소가 있고 그 요소에 string이 오직 하나 있다면 
# 그냥 anchor에서 string method 실행하면 됌
pages = []
for link in links[:-1]:
    pages.append(int(link.string))  # 그 안에 string만 가져오게 설정
                                    # int로 변경
max_page = pages[-1]
```

```python
for n in range(max_page):
    print("start={}".format(n*50))
```

### function
- 위의 코드들을 하나의 함수로 만듬
- 페이지 구조가 바뀌어서 다름
- #2.5 댓글들 보면 많은 사람들이 해결법 올려줌 (참고)

```python
import requests
from bs4 import BeautifulSoup

LIMIT = 50
URL = "https://kr.indeed.com/jobs?q=python&limit={}".format(LIMIT)

def extract_indeed_pages():
    result = requests.get(URL)
    soup = BeautifulSoup(result.text, "html.parser")
    pagination = soup.find("div", {"class":"pagination"})

    links = pagination.find_all('a')
    pages = []
    for link in links[:-1]:
        pages.append(int(link.string))
    
    max_page = pages[-1]
    return max_page

def extract_indeed_jobs(last_page):
    jobs = []
#     for page in range(last_page):
    result = requests.get("{}&start={}".format(URL,0*LIMIT))
    soup = BeautifulSoup(result.text, "html.parser")
    # results : 각각 일자리 list
    results = soup.find_all("div", {'class':'jobsearch-SerpJobCard'})
    for result in results:
        title = result.find('h2', {'class':'title'}).find('a')['title']
        company = result.find('span', {'class':'company'})
        company_anchor = company.find('a')
        if  company_anchor is not None:
            company = str(company_anchor.string)
        else:
            company = str(company.string)
        company = company.strip() # 빈칸이 많이 나와 양쪽에 빈칸 삭제
    return jobs
```

### main

```python
last_indeed_pages = extract_indeed_pages()
print(last_indeed_pages)

indeed_jobs = extract_indeed_jobs(last_indeed_pages)
print(extract_indeed_jobs)
```
